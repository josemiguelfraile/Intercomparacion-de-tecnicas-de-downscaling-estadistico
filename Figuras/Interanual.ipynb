# Representación Interanual

En este notebook se llevara a cabo la representación gráfica interanual para dos modelos(GLM16 y CNN).

- Cargamos las librerias necesarias para la elaboración del notebook.

```r
library(loadeR)
library(magrittr)
require(transformeR)
require(downscaleR)
require(visualizeR)
require(climate4R.value)
require(downscaleR.keras)
library(ggplot2)
```

- Cargamos las variables predictoras y predictando.

```r
y <- load('../Data/y_sam.rda') %>% get()
x <- load('../Data/x_SAM.rda') %>% get()
```

- Se cargan los modelos.

```r
glm1 <- load('../GLM_1.rda') %>% get()
glm4 <- load('../GLM_4.rda') %>% get()
glm16 <- load('../GLM_16.rda') %>% get()
nnss <- load('../nn-ss1.rda') %>% get()
nnss4 <- load('../nn-ss4.rda') %>% get()
nnss16 <- load('../nn-ss16.rda') %>% get()
cnnss <- load('../cnn-ss.rda') %>% get()
analogos <- load('../Analogos.rda') %>% get()
```

- Se carga el archivo con las regiones.

```r
regiones <- read.csv('../regiones_VALUE_SAM.csv')
```

- Agrupamos los modelos por valores anuales.

```r
aggr.pars <- list(FUN = "mean", na.rm=TRUE)
```

```r
obs <- aggregateGrid(y, aggr.y= aggr.pars)
Glm1 <- aggregateGrid(glm1, aggr.y= aggr.pars)
Glm4 <- aggregateGrid(glm4, aggr.y= aggr.pars)
Glm16 <- aggregateGrid(glm16, aggr.y= aggr.pars)
Nnss <- aggregateGrid(nnss, aggr.y= aggr.pars)
Nnss4 <- aggregateGrid(nnss4, aggr.y= aggr.pars)
Nnss16 <- aggregateGrid(nnss16, aggr.y= aggr.pars)
Cnnss <- aggregateGrid(cnnss, aggr.y= aggr.pars)
Analogos <- aggregateGrid(analogos, aggr.y= aggr.pars)
```

- Representamos la serie interanual para los modelos GLM16 y CNN divididos en regiones. A este gráfico se le añadira una región sombreada que representa media desviación tipica para cada uno de los modelos.

```r
# Gráfico interanual con las sombras
#pdf('Plot_Interanual.pdf',width = 4,height = 4) Esta linea es para guradar la imagen como pdf.

lapply(c('SESA','AR','AP','NC','CSC'), FUN = function(reg){
    region <- subset(regiones, regiones$Region == reg)
    OBS <- subsetGrid(obs, station.id = region[,1])
    GLM1 <- subsetGrid(Glm1, station.id = region[,1])
    GLM4 <- subsetGrid(Glm4, station.id = region[,1])
    GLM16 <- subsetGrid(Glm16, station.id = region[,1])
    NN <- subsetGrid(Nnss, station.id = region[,1])
    NN4 <- subsetGrid(Nnss4, station.id = region[,1])
    NN16 <- subsetGrid(Nnss16, station.id = region[,1])
    CNN <- subsetGrid(Cnnss, station.id = region[,1])
    
    #Observación
    plot(as.Date(OBS$Dates$start),rowMeans(OBS$Data,na.rm=TRUE),type='l',xlab = "año",main = reg,
                                ylab = "temp", col= '#030303')
    
    #Dibujo de GLM16(Azul)
    col3 <- rgb(0, 0, 1, alpha = 0.1)
    polygon(c(as.Date(OBS$Dates$start), rev(as.Date(OBS$Dates$start))),
            c(rowMeans(GLM16$Data,na.rm=TRUE) -0.5*sd(rowMeans(GLM16$Data,na.rm=TRUE)),
              rev(rowMeans(GLM16$Data,na.rm=TRUE) +0.5*sd(rowMeans(GLM16$Data,na.rm=TRUE)))), col = col3,border=col3)
    #lines(x = as.Date(OBS$Dates$start), rowMeans(GLM16$Data,na.rm=TRUE) -sd(rowMeans(GLM16$Data,na.rm=TRUE)), col = "#1E90FF")
    #lines(x = as.Date(OBS$Dates$start), rowMeans(GLM16$Data,na.rm=TRUE) +sd(rowMeans(GLM16$Data,na.rm=TRUE)), col="#1E90FF")
    lines(as.Date(GLM16$Dates$start),rowMeans(GLM16$Data,na.rm=TRUE), col = '#1E90FF')
    
    
    #Dibujo de CNN(Rojo)
    col2= rgb(1, 0, 0, alpha = 0.1)
    polygon(c(as.Date(OBS$Dates$start), rev(as.Date(OBS$Dates$start))),
            c(rowMeans(CNN$Data,na.rm=TRUE) -0.5*sd(rowMeans(CNN$Data,na.rm=TRUE)),
              rev(rowMeans(CNN$Data,na.rm=TRUE) +0.5*sd(rowMeans(CNN$Data,na.rm=TRUE)))), col = col2,border=col2)
    #lines(x = as.Date(OBS$Dates$start), rowMeans(NN16$Data,na.rm=TRUE) -0.5*sd(rowMeans(NN16$Data,na.rm=TRUE)), col = "#FF6A6A")
    #lines(x = as.Date(OBS$Dates$start), rowMeans(NN16$Data,na.rm=TRUE) +0.5*sd(rowMeans(NN16$Data,na.rm=TRUE)), col="#FF6A6A")
    
    
    lines(as.Date(CNN$Dates$start),rowMeans(CNN$Data,na.rm=TRUE), col = '#EE2C2C')
    
    
    
    
    legend(x = "topright",
       inset = c (-0.5, 0),
       legend = c("Observacion","GLM16","CNN"), 
       cex = 0.25,
       lty = c(1, 1),          
       col = c('#030303', '#FF6A6A'),          
       lwd = 0.5,
       xpd = TRUE,
       merge = TRUE,
       horiz = FALSE)
    
})
#dev.off()   Esta linea es para guradar la imagen como pdf.
```
